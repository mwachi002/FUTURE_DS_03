```python
import pandas as pd
```

```python
df = pd.read_csv('/content/student_feedback.csv')
```

```python
# Display first 5 rows
display(df.head())
```

```python
# Missing values per column
print(df.isnull().sum())

# Number of duplicate rows
print(df.duplicated().sum())

# DataFrame information and data types
df.info()
```

```python
# Descriptive statistics
rating_columns = df.columns.drop('Student ID')
descriptive_stats = df[rating_columns].describe()
print("Descriptive Statistics for Numerical Rating Columns:")
display(descriptive_stats)
```

```python
import matplotlib.pyplot as plt

average_ratings = descriptive_stats.loc['mean']

plt.figure(figsize=(12, 6))
average_ratings.plot(kind='bar', color='skyblue')
plt.title('Average Student Ratings Across Categories')
plt.xlabel('Rating Categories')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
